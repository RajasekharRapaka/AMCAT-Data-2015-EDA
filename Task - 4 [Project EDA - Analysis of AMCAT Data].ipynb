```python
from collections import Counter
from fuzzywuzzy import fuzz

# Assuming 'city' is the Series containing the city counts

# Define a function to calculate similarity based on common letters
def common_letters_similarity(word1, word2):
    # Calculate the intersection of letter counts between the two words
    common_letters_count = sum((Counter(word1) & Counter(word2)).values())
    # Normalize the similarity score by the maximum possible number of common letters
    max_possible_common_letters = min(len(word1), len(word2))
    similarity = common_letters_count / max_possible_common_letters
    return similarity

# Define a function to normalize city names based on common letters similarity
def normalize_city(city_name, city_mapping):
    best_match, best_score = None, 0
    for mapped_city, aliases in city_mapping.items():
        for alias in aliases:
            similarity = common_letters_similarity(city_name, alias)
            if similarity > best_score:
                best_match, best_score = mapped_city, similarity
    if best_score >= 0.6:  # Adjust the similarity threshold as needed
        return best_match
    else:
        return city_name

# Normalize city names based on common letters similarity
city_mapping = {
    'bengaluru': ['bangalore', 'bengaluru', 'blore', 'bnglr', 'Asifabadbanglore'],
    'noida': ['noida'],
    'hyderabad': ['hyderabad', 'hyd'],
    'pune': ['pune'],
    'kolkata' : ['kolkata'],
    'muzaffarpur' : ['muzzafarpur'],
    'mumbai' : ['mumbai', ' mumbai'],
    'tirupati' : ['tirupathi'],
    'lucknow' : ['lucknow'],
    'vizag' : ['vsakhapttnam']
    # Add more city mappings as needed
}

# Convert 'JobCity' column values to lowercase
data['JobCity'] = data['JobCity'].str.lower()

# Apply normalization function to 'JobCity' column
data['JobCity'] = data['JobCity'].map(lambda x: normalize_city(x, city_mapping))

# Print the unique values in the 'JobCity' column to verify changes
print(data['JobCity'].unique())
```

# 1 sample T-test to verify the claim

Null Hypothesis: mu = 250k
Alternate Hypothesis : mu != 250k

```python
from scipy import stats as st
from scipy.stats import chi2_contingency as cst
pv = st.ttest_1samp(new['Salary'],popmean=250000)[1]
### for a 95% confidence interval,my p- value should be >0.05 to claim the null hypothesis
if pv < 0.05:
    print('We reject the null hypothesis and Average salary is not equal to 250k')
else:
    print('We fail to reject null hypothesis and Avergae salary is equal to 250k')
```
